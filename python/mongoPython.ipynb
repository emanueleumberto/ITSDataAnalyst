```python
from bson import Regex
import pymongo
connectionString = "mongodb+srv://admin:root@cluster0.7opl1ou.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0"

cluster = pymongo.MongoClient(connectionString)
database = cluster.DataMongoPython
users = database['users']

# Filtrare dati nella collection
# nel metodo find[find_one] il primo parametro è un oggetto di tipo query
# possiamo utilizzarlo per filtrare i risultati della collection 
list = users.find({"city": "Roma"})
# print(list)
# for o in list:
#    print(o)

# Seleziono tutti gli oggetti in cui la proprietà city è 
# contenuta($in) in un set di valori
filterList = users.find({"city": {"$in": ["Roma", "Milano", "Napoli"]}})
# for o in filterList:
#    print(o)

# Seleziono tutti gli oggetti in cui la proprietà age è 
# maggiore($gt) di un determinato valore
filterAge = users.find({"age": {"$gt": 30}})
# for o in filterAge:
#    print(o)
    
# Seleziono tutti gli oggetti in cui la proprietà age è 
# minore($lt) di un determinato valore
filterAge = users.find({"age": {"$lt": 30}})
# for o in filterAge:
#    print(o)

# Seleziono tutti gli oggetti in cui la proprietà name AND la 
# proprietà city hanno un determinato valore
filterAnd = users.find({"name": "Mario", "city": "Roma"})
# for o in filterAnd:
#    print(o)

# Seleziono tutti gli oggetti in cui la proprietà name OR la 
# proprietà city hanno un determinato valore
filterOr = users.find({"$or": [{"name": "Giuseppe"}, {"city": "Roma"}]})
# for o in filterOr:
#    print(o)

# E' anche possibile utilizzare le espressioni regolari come filtro
# In questo esempio abbiamo selezionato tutti gli oggetti che hanno
# un cognome che comincia per V
filterRegex = users.find({"lastname": {"$regex": "^V"}})
# for o in filterRegex:
#    print(o)

# Utilizzando un metodo sort() ho la possibilità di ordinare i dati
# in ordine crescente o decrescente
# sort(field, 1) -> Crescente
# sort(field, -1) -> Decrescente
list = users.find().sort('lastname', -1)
# for o in list:
#    print(o['name'], o['lastname'])

# Per limitare i risultati di una ricerca in MongoDB possiamo utilizzare
# il metodo limit(limit: int) come paramentro, inserire il numero di 
# risultati che si vuole ottenere. 
# Con il metodo skip(skip: int) posso indicare il numero di risultati
# da escludere dalla richiesta
list = users.find().limit(3).skip(3)
for o in list:
    print(o['name'], o['lastname'])
```

```python
import pymongo
connectionString = "mongodb+srv://admin:root@cluster0.7opl1ou.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0"

cluster = pymongo.MongoClient(connectionString)
database = cluster.DataMongoPython
users = database['users']

# Modificare un dato nella collection -> UPDATE di sql
# update_one(filter, update)
# update_many()

# Il primo parametro del metodo update_one è un oggetto di tipo query
# che definisce quale elemento aggiornare, il secondo paramentro
# è un oggetto con le modifiche da applicare all'elemento selezionato
result = users.update_one({"city": "Roma"}, {"$set": {"name": "Paolo", "lastname": "Gialli"}})
print(result.modified_count, "Modified")

# Il primo parametro del metodo update_many è un oggetto di tipo query
# che definisce quale elemento aggiornare, il secondo paramentro
# è un oggetto con le modifiche da applicare a tutti gli elementi selezionati dalla query
result = users.update_many({"lastname": {"$regex": "^V"}}, {"$set": {"age": 99}})
print(result.modified_count, "Modified")

```

```python
import pymongo
connectionString = "mongodb+srv://admin:root@cluster0.7opl1ou.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0"

cluster = pymongo.MongoClient(connectionString)
database = cluster.DataMongoPython
users = database['users']

# Eliminare un dato nella collection -> DELETE di sql
# Il parametro del metodo delete_one è un oggetto di tipo query
# che definisce quale elemento eliminare nel database
result = users.delete_one({"city": "Roma"})
print(result.deleted_count, "deleted!")

# Il parametro del metodo delete_many è un oggetto di tipo query
# che definisce tutti gli elementi da eliminare nel database
result = users.delete_many({"city": "Milano"})
print(result.deleted_count, "deleted!")

```
