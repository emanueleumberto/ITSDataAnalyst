```python
# Series e DataFrame
import pandas as pd

#Series
pds = pd.Series([1,2,3,4,5])
# print(pds)
print(pds[2])

pdsi = pd.Series([1,2,3,4,5], index=list('abcde'))
# print(pdsi)
print(pdsi['c'])
```

```python
# loc[] -> usa le etichette
# iloc[] -> usa gli indici (slicing)

pdsi = pd.Series([1,2,3,4,5], index=['AB', 'EF','CD',  'GH', 'IL'])

# print(pdsi.iloc[3])
# print(pdsi.loc["CD"])

#print(pdsi.iloc[1:3])
print(pdsi.loc["AB": "CD"])
```
